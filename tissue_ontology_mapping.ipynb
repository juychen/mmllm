### How many tracks are there per output type?

````python
# Count human tracks
human_tracks = (
    dna_model.output_metadata(dna_client.Organism.HOMO_SAPIENS)
    .concatenate()
    .groupby('output_type')
    .size()
    .rename('# Human tracks')
)

# Count mouse tracks
mouse_tracks = (
    dna_model.output_metadata(dna_client.Organism.MUS_MUSCULUS)
    .concatenate()
    .groupby('output_type')
    .size()
    .rename('# Mouse tracks')
)

pd.concat([human_tracks, mouse_tracks], axis=1).astype(pd.Int64Dtype())
````

````python
from alphagenome.data.genome import Interval
import matplotlib.pyplot as plt
import numpy as np
````

````python
df_dmr_hip_5mc = df_dmr_hip[df_dmr_hip['mod'] == '5mC']
df_dmr_hip_5mc.sort_values(by=['score'], inplace=False,ascending=False)
````

````python
track1s= []
track2s = []
for index,row in df_dmr_hip_5mc.iterrows():
    chrom = f"chr{row['chr']}"
    start = row['start']
    end = row['end']
    name = row['dmr']
    #print(f"Processing {chrom}:{start}-{end}")
    # output = dna_model.predict_region(
    #     chrom=chrom,
    #     start=start,
    #     end=end,
    #     requested_outputs=[dna_client.OutputType.CHIP_HISTONE],
    #     ontology_terms=['UBERON:0002305'],  # Lung.
    # )
    #5mC|CG_11:12025411-12026932
# pad interval to 2048
# 原始信息
    chrom = chrom
    orig_start = start
    orig_end = end
    target_length = 16384
    name = name

    # 计算当前长度和所需 padding
    current_len = orig_end - orig_start
    pad_total = target_length - current_len

    # 计算左右两侧的扩展量（保持中心对齐）
    pad_left = pad_total // 2
    pad_right = pad_total - pad_left

    # 创建 Padding 后的新 Interval
    # 注意：在实际应用中，建议检查 start 是否小于 0
    new_start = max(0, orig_start - pad_left)
    new_end = new_start + target_length

    interval_padded = Interval(chrom, start=new_start, end=new_end,name=name)
    output = dna_model.predict_interval(
        
        interval=interval_padded,
        organism=dna_client.Organism.MUS_MUSCULUS,
        requested_outputs=[dna_client.OutputType.CHIP_HISTONE],
            ontology_terms=['UBERON:0002305'],  # Lung.

    )
    # plot the value of the first track
    # x start from satrt-16384/2 to end+16384/2, step = 128
    # x = np.arange(0 - target_length // 2, 0 + target_length // 2, 128)
    # plt.plot(x, output.chip_histone.values[:,0], label='Track 1 H3K27ac')
    # plt.plot(x, output.chip_histone.values[:,1], label='Track 2 H3K4me3')
    # plt.title('Predicted CHIP_HISTONE signal')
    # plt.xlabel('Position (bp)')
    # plt.ylabel('Signal')
    # plt.legend()
    # plt.show()
    # print(f"Predicted CHIP_HISTONE signal shape: {output.chip_histone.values.shape}")
    track1s.append(output.chip_histone.values[:,0])
    track2s.append(output.chip_histone.values[:,1])
````
